##### __(d) Now, Calvin has watched the movie with the ID 1502397 and we want to recommend movies with a similar plot. Remove all punctuation marks (".", ",", "(", ")", "-", "&"), convert the content to lower-case, and split the content of the plot attribute by whitespaces into terms. Then remove all stopwords by using the stopwords_english.txt file. Then calculate the Jaro Winkler distance (make use of an existing library) between the pre-processed plot of the watched movie and the pre-processed plots of all other movies. Print the top 3 movies based on the calculated  Jaro Winkler distance and the rating of the movie.  [4 points].__ 

```python
import pandas as pd
import numpy as np
from strsimpy.jaro_winkler import JaroWinkler

df = pd.read_csv('movies.csv')
df = df.sort_values('rating', ascending=False)


def pp_plot(plot):
    punc = [".", ",", "(", ")", "-", "&"]
    for p in punc:
        plot = plot.replace(p, '')
    plot = plot.lower().split(' ')
    with open('stopwords_english.txt', encoding='utf-8') as file:
        stop_words = file.readlines()
        stop_words = [s.replace("\n", "") for s in stop_words]

        # remove all stop words from plot
        for i in range(len(plot)):
            if plot[i] in stop_words:
                plot[i] = ""
        
        plot = list(filter(None, plot))
            
    return plot

#print(pp_plot('Set within a year after the events of Batman Begins (2005), Batman jan'))

original_plot = df.loc[df['id'] == 1502397]['plot'].values[0]
df = df.loc[df['plot'] != original_plot]

#
df = df.loc[df['id'] != 1502397]
df = df.loc[df['id'] != 9999999]


jarowinkler = JaroWinkler()

for i, row in df.iterrows():
    df.at[i,'diff'] = jarowinkler.similarity(pp_plot(original_plot), pp_plot(row['plot']))

df = df.sort_values('diff', ascending=False) 
df.head(3)

#print(jarowinkler.similarity(pp_plot(original_plot), pp_plot(original_plot)))
```

### Task 02: Implementation of collaborative filtering recommenders [total: 12 points]. 
The ratings.csv file contains a dataset extracted from IMDB with ratings for movies in a range from 1 to 5 from several users. The meta-data for the movies is contained in the movies.csv file. Use the movies.csv file for printing the movie titles in the results. Implement a user-based and a item-based collaborative filtering recommendation engine like show in the lecture. ##### __(a)  Print the top 3 recommended movies for the users 'Delora', 'Ruth', 'Heather', 'Ronald', and 'Lakisha' in the dataset for the user-based version. ##### __(c)  Print the top 3 recommended movies for the users 'Harold', 'Alfredo', 'Jerry', 'Monica', and 'Gail' in the dataset for the item-based version.  We also use the threshold version for choosing similar items. The threshold for choosing the similar items needs to be set to 0.35. [4 points]__ 

```python
import pandas as pd
import seaborn as sns
import numpy as np

df_r = pd.read_csv('ratings.csv')
df_m = pd.read_csv('movies.csv')
#print(df_r.describe())

#delete data that does not fit 
df_r = df_r.loc[df_r['rating'] != 333]

#create similarity matrix
similarity_item = pd.DataFrame({'id_1': [], 'id_2': [], 'similarity': []})

# calculate similarity based on two vector (lists)
def similarity(item_v1, item_v2):
    dividend = sum(item_v1 * item_v2)
    divisor = np.linalg.norm(item_v1, 2) * np.linalg.norm(item_v2, 2)
    return dividend / divisor

# test similarity from lecture -> 0.27
# print(similarity(np.array([4,5,4,0,0,0]), np.array([3,0,0,3,4,0])))

#create matrix
movie_matrix_user = df_r.pivot_table(index='user', columns='id', values='rating')
# replace NaN
movie_matrix_user = movie_matrix_user.fillna(0)


for in_row, row in enumerate(movie_matrix_user):
    
    base_column = movie_matrix_user.iloc[:, in_row]
    base_column_id = movie_matrix_user.columns[in_row]
   
    for in_col, column in enumerate(movie_matrix_user):
   
        column_curr = movie_matrix_user[column]    
        similarity_item = similarity_item.append({'id_1': base_column_id, 
                                              'id_2': movie_matrix_user.columns[in_col], 
                                              'similarity': similarity(column_curr,base_column)}, ignore_index=True) 
    
# set similarity to 0 for the item to itself
similarity_item.loc[similarity_item['id_1'] == similarity_item['id_2'], ['similarity']] = 0
#df =similarity_item.loc[similarity_item['id_1'] == similarity_item['id_2']]
print(similarity_item)
```

```python
#threshold 0.35 >= ? or >
similarity_item = similarity_item[similarity_item['similarity'] > 0.35]
#print(similarity_item)

def recommendationItemBased(user, n):
    recommendation_items = pd.DataFrame({'id': [], 'recommendation': [], 'title': []})

    #all rated items from user
    items_user_rating = df_r.loc[df_r['user'] == user]
    
    items_probably_not_rated = pd.DataFrame({'id_rate': [], 'rate': [], 'id_sim': [], 'similarity': []})
    
    for index, row in items_user_rating.iterrows():        
        item = row['id']
        rating = row['rating']
        similarity_row_items = similarity_item.loc[similarity_item['id_1'] == item]

        for i, sim_item in similarity_row_items.iterrows():           
            items_probably_not_rated = items_probably_not_rated.append({'id_rate':item,
                                                                        'rate': rating,
                                                                        'id_sim': sim_item['id_2'],
                                                                        'similarity': sim_item['similarity']}, ignore_index=True) 
     
    #print(items_probably_not_rated)
    
    #select only unrated items from user
    items_not_rated = items_probably_not_rated.loc[~items_probably_not_rated.id_sim.isin(items_user_rating.id)]
    
    #print(items_not_rated)
    
    #calculate recommendation for user not rated items
    for item in set(items_not_rated.id_sim):
        dividend = 0
        divisor = 0
        calc_recommendation = items_not_rated.loc[items_not_rated['id_sim'] == item]
        
        #calc for one item recommendation
        for in_calc, row_calc in calc_recommendation.iterrows():
            dividend += (row_calc['rate'] * row_calc['similarity'])
            divisor += row_calc['similarity']
        
        recommendation = dividend/divisor   
        recommendation_items = recommendation_items.append({'id': item, 
                                                            'recommendation': recommendation,
                                                           'title': df_m.loc[df_m['id'] == item]['title'].values[0]}, 
                                                           ignore_index=True)
    
    recommendation_items = recommendation_items.sort_values('recommendation', ascending=False)
    return recommendation_items[:n]

print(recommendationItemBased('Harold', 3))
print(recommendationItemBased('Alfredo', 3))
print(recommendationItemBased('Jerry', 3))
print(recommendationItemBased('Monica', 3))
print(recommendationItemBased('Gail', 3))

#print(recommendationItemBased('Chantal', 15))
```

##### __(d) Print the user, which has the movie 'Pomerania Courier' as best recommendation for the item-based version from Task 02 c)? [1 point]__ 

```python
import pandas as pd
import seaborn as sns
import numpy as np

df_r = pd.read_csv('ratings.csv')
df_m = pd.read_csv('movies.csv')

id_movie = df_m.loc[df_m['title'] == 'Pomerania Courier']['id'].values[0]

names = df_r['user'].unique()
users_rec = pd.DataFrame({'user': [], 'recommendation': [], 'title': []})

for name in names:
    reccomendations = recommendationItemBased(name, 30)
    for idx, row in reccomendations.iterrows():    
        if row['id'] == id_movie:
            users_rec =  users_rec.append({'user': name,
                                          'recommendation': row['recommendation'],
                                          'title': row['title']}, ignore_index=True)
    
users_rec = users_rec.sort_values('recommendation', ascending=False)
users_rec.head()
```
