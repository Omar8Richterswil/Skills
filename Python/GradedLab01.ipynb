### Task 02: Implementation of collaborative filtering recommenders [total: 12 points]. 
The ratings.csv file contains a dataset extracted from IMDB with ratings for movies in a range from 1 to 5 from several users. The meta-data for the movies is contained in the movies.csv file. ##### __(a)  Print the top 3 recommended movies for the users 'Delora', 'Ruth', 'Heather', 'Ronald', and 'Lakisha' in the dataset for the user-based version.  We use the threshold version for choosing similar users. The threshold for choosing the similar users needs to be set to 0.25. [5 points]__ 

```python
import pandas as pd
import numpy as np
import operator

np.seterr(divide='ignore', invalid='ignore')

df_ratings = pd.read_csv('ratings.csv')
df_movies = pd.read_csv('movies.csv')

# Cleanup Data
df_movies = df_movies[df_movies.id != 9999999]
df_ratings = df_ratings.drop(df_ratings[df_ratings.rating < 1].index)
df_ratings = df_ratings.drop(df_ratings[df_ratings.rating > 5].index)

cos_sim = lambda a,b: np.dot(a,b.T)/(np.linalg.norm(a) * np.linalg.norm(b))

def sim(user, rating_matrix):   
    user_rating = rating_matrix[rating_matrix.index == user]
    other_user_rating = rating_matrix[rating_matrix.index != user]

    user_rating = user_rating.loc[:, (user_rating.isnull().sum(axis=0) <= 0)]
    other_user_rating = other_user_rating[user_rating.columns]
    
    similarities = pd.DataFrame(np.zeros((len(other_user_rating.index))), other_user_rating.index, columns=['similarity'])
    
    
    for idx, row in other_user_rating.iterrows():
        other_user = pd.DataFrame(row).transpose()
        other_user = other_user.loc[:, (other_user.isnull().sum(axis=0) <= 0)]
        
        common_cols = list(set(user_rating.columns).intersection(other_user.columns))
        
        if len(common_cols) == 0:
            similarities.at[row.name, 'similarity'] = 0
        else:
            sim = cos_sim(user_rating[common_cols], other_user[common_cols])
            similarities.at[row.name, 'similarity'] = sim[0][0]
    
    index_similarity = pd.DataFrame(similarities, other_user_rating.index, columns=['similarity'])
    index_similarity.index.name = 'user'
    index_similarity = index_similarity.sort_values('similarity', ascending=False)
    index_similarity = index_similarity.drop(index_similarity[index_similarity.similarity <= 0.25].index)
    
    return index_similarity

def recommend(user, similar_users, rating_matrix, k=3):
    
    # Get ratings from similar users
    similar_user_rating = rating_matrix.loc[similar_users.index]
    
    # Load current user ratings
    current_user_rating = rating_matrix.loc[[user]].transpose()
    current_user_rating.columns = ['rating']
    current_user_rating = current_user_rating.dropna()
    
    all_movie_indices = similar_user_rating.columns
    watched_movies = current_user_rating.index.values.tolist()
    
    # Get movies that are not watched by the user 
    unwatched_moves = list(set(all_movie_indices) - set(watched_movies))
    
    recommendations = pd.DataFrame(columns=['movie','recommendation'])
    for idx, movie in enumerate(unwatched_moves):
        ratings_by_users = similar_user_rating[movie].dropna()
        
        val_0 = 0
        for user, rating in ratings_by_users.iteritems():
            sim = similar_users.loc[user][0]
            val_0 += sim * rating
        
        val_1 = 1
        for user, rating in ratings_by_users.iteritems():
            sim = similar_users.loc[user][0]
            val_1 += sim
            
        rec = pd.DataFrame([[movie,(val_0 / val_1)]], columns=['movie','recommendation'])
        recommendations = pd.concat([recommendations, rec])
    
    recommendations = recommendations.sort_values('recommendation', ascending=False)
    return recommendations[:k]


rating_matrix = df_ratings.pivot_table(index='user', columns='id', values='rating')
sim_matrix = pd.DataFrame(index=rating_matrix.index)

for user in sim_matrix.index:
    print('.', end = '' )
    sim_matrix[user] = sim(user, rating_matrix)

names = ['Delora', 'Ruth', 'Heather', 'Ronald', 'Lakisha']
for name in names:
    print('\n', name)
    reccomendations = recommend(name, sim(name, rating_matrix), rating_matrix)
    for idx, row in reccomendations.iterrows():
        movie = df_movies.loc[df_movies['id'] == row['movie']]['title'].values[0]
        print('  -', movie, row['movie'])
        print('   ', row['recommendation'])
```

##### __(c)  Print the top 3 recommended movies for the users 'Harold', 'Alfredo', 'Jerry', 'Monica', and 'Gail' in the dataset for the item-based version.  We also use the threshold version for choosing similar items. The threshold for choosing the similar items needs to be set to 0.35. [4 points]__ 

```python
import pandas as pd
import seaborn as sns
import numpy as np

df_r = pd.read_csv('ratings.csv')
df_m = pd.read_csv('movies.csv')

# Cleanup Data
df_movies = df_movies[df_movies.id != 9999999]
df_ratings = df_ratings.drop(df_ratings[df_ratings.rating < 1].index)
df_ratings = df_ratings.drop(df_ratings[df_ratings.rating > 5].index)

def similarity(item_v1, item_v2):
    dividend = sum(item_v1 * item_v2)
    divisor = np.linalg.norm(item_v1, 2) * np.linalg.norm(item_v2, 2)
    return dividend / divisor

def recommendationItemBased(user, n):
    recommendation_items = pd.DataFrame(columns=['id','recommendation','title'])

    #all rated items from user
    items_user_rating = df_r.loc[df_r['user'] == user]
    
    items_probably_not_rated = pd.DataFrame(columns=['id_rate', 'rate', 'id_sim', 'similarity'])
    
    for index, row in items_user_rating.iterrows():     
        item = row['id']
        rating = row['rating']
        similarity_row_items = similarity_item.loc[similarity_item['id_1'] == item]

        for i, sim_item in similarity_row_items.iterrows():           
            item = pd.DataFrame([[item, rating, sim_item['id_2'], sim_item['similarity']]], 
                                 columns=['id_rate', 'rate', 'id_sim', 'similarity'])
            items_probably_not_rated = pd.concat([items_probably_not_rated, item])
         
    #select only unrated items from user
    items_not_rated = items_probably_not_rated.loc[~items_probably_not_rated.id_sim.isin(items_user_rating.id)]
    
    #print(items_not_rated)
    
    #calculate recommendation for user not rated items
    for item in set(items_not_rated.id_sim):
        dividend = 0
        divisor = 0
        calc_recommendation = items_not_rated.loc[items_not_rated['id_sim'] == item]
        
        #calc for one item recommendation
        for in_calc, row_calc in calc_recommendation.iterrows():
            dividend += (row_calc['rate'] * row_calc['similarity'])
            divisor += row_calc['similarity']
        
        recommendation = dividend/divisor   
        rec_item = pd.DataFrame([[item, recommendation, df_m.loc[df_m['id'] == item]['title'].values[0]]],
                                columns=['id','recommendation','title'])
        recommendation_items = pd.concat([recommendation_items, rec_item])
    
    recommendation_items = recommendation_items.sort_values('recommendation', ascending=False)
    return recommendation_items[:n]

#create similarity matrix
similarity_item = pd.DataFrame(columns=['id_1','id_2','similarity'])
#create rating matrix
movie_matrix_user = df_r.pivot_table(index='user', columns='id', values='rating')
# replace NaN
movie_matrix_user = movie_matrix_user.fillna(0)


for in_row, row in enumerate(movie_matrix_user):
    print('.', end = '' )
    
    base_column = movie_matrix_user.iloc[:, in_row]
    base_column_id = movie_matrix_user.columns[in_row]
   
    for in_col, column in enumerate(movie_matrix_user):
   
        column_curr = movie_matrix_user[column]
        item = pd.DataFrame([[base_column_id, movie_matrix_user.columns[in_col],
                             similarity(column_curr,base_column)]], columns=['id_1','id_2','similarity']) 
        similarity_item = pd.concat([similarity_item, item])

#threshold 0.35 
similarity_item = similarity_item[similarity_item['similarity'] > 0.35]

names = ['Harold', 'Alfredo', 'Jerry', 'Monica', 'Gail']
for name in names:
    print('\n----------------------------------------------------')
    print('Recommendations for', name)
    print(recommendationItemBased(name, 3))
```
